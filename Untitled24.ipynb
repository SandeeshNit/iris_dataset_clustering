```python
#From the given ‘Iris’ dataset, predict the optimum number of clusters and represent it visually.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris = datasets.load_iris()
iris = pd.DataFrame(data=iris.data,columns=iris.feature_names)
```

```python
#now find the optimum number of clusters for classification
x = iris.iloc[:,[0,1,2,3]].values

from sklearn.cluster import KMeans
wcss = []
for i in range (1,11):
    kmeans = KMeans(n_clusters= i)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
#plot the result 
plt.plot(range(1,11),wcss)
plt.title("The Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters= 3)   # as we can determine the optimum number of clusters from the plot above
y_kmeans = kmeans.fit_predict(x)
```

```python
#Visualizing
plt.scatter(x[y_kmeans == 0 , 0], x[y_kmeans == 0,1], c = 'red' , label = 'Iris_setosa')
plt.scatter(x[y_kmeans == 1 , 0], x[y_kmeans == 1,1],c = 'blue' , label = 'Iris_Versicolour')
plt.scatter(x[y_kmeans == 2 , 0], x[y_kmeans == 2,1], c = 'green' , label = 'Iris_Virginica')
#Plotting the centroids
plt.scatter(kmeans.cluster_centers_[: , 0], kmeans.cluster_centers_ [:,1], c = 'yellow' , label = 'Centroids')
plt.legend(loc=4)
```
